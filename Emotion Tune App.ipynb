```python
import sqlite3

def create_database():
    conn = sqlite3.connect('emotion_tune.db')  # Replace with your desired database name
    cursor = conn.cursor()

    with open('emotion_tune_database.sql', 'r') as f:
        sql_script = f.read()

    cursor.executescript(sql_script)
    conn.commit()
    conn.close()

if __name__ == "__main__":
    create_database()
```
